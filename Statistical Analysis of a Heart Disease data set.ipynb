# Key information about the dataframe 
- Age: Patients Age in years (Numeric)
- Sex: Gender (Male : 1; Female : 0) (Nominal)
- cp: Type of chest pain experienced by patient. This term categorized into 4 category where 0 typical angina, 1 atypical angina, 2 non- anginal pain, 3 asymptomatic (Nominal)
- trestbps: patient's level of blood pressure at resting mode in mm/HG (Numerical)
- chol: Serum cholesterol in mg/dl (Numeric)
- fbs: Blood sugar levels on fasting > 120 mg/dl represents as 1 in case of true and 0 as false (Nominal)
- restecg: Result of electrocardiogram while at rest are represented in 3 distinct values where 0 : Normal 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of >
- 0.05 mV) 2: showing probable or definite left ventricular hypertrophyby Estes' criteria (Nominal)
- thalach: Maximum heart rate achieved (Numeric)
- exang: Angina induced by exercise 0 depicting NO 1 depicting Yes (Nominal)
- oldpeak: Exercise induced ST-depression in relative with the state of rest (Numeric)
- slope: ST segment measured in terms of slope during peak exercise where 0: up sloping; 1: flat; 2: down sloping(Nominal)
- ca: The number of major vessels (0â€“3)(nominal)
- thal: A blood disorder called thalassemia where 0: NULL 1: normal blood flow 2: fixed defect (no blood flow in some part of the heart) 3: reversible defect (a blood flow is observed but it is not normal(nominal)
- target: It is the target variable which we have to predict 1 means patient is suffering from heart disease and 0 means patient is normal.

```python
#"age"_calculations

#5 Number summary
print("Below are calculations for the Age column:")
print()
print("The 5 number summary:")
print()
minValue = heartDisease_df["age"].min()
print("The minimmum value is", minValue)
quartile1 = p(heartDisease_df["age"],25)
print("The 25th percentile is",quartile1)
quartile2 = heartDisease_df["age"].median()
print("The median is",quartile2)
quartile3 = p(heartDisease_df["age"],75)
print("The 75th percentile is",quartile3)
maxValue = heartDisease_df["age"].max()
print("The maximum value is",maxValue)
print()
print("Other stats calculations:")
print()
modeValue = heartDisease_df["age"].mode()
print("The mode is", modeValue)
averageValue = heartDisease_df["age"].mean()
print("The mean is ", averageValue)
rangeValue = (heartDisease_df["age"].max() -heartDisease_df["age"].min())
print("The range is", rangeValue)
IQRvalue = quartile3 - quartile1
print("The interquartile range is", IQRvalue)
variationValue = np.var(heartDisease_df["age"])
print("The variation is", variationValue)
stdVariationValue = np.std(heartDisease_df["age"])
print("The standard deviation is", stdVariationValue)
```
